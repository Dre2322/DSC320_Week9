```python
# 1. Finding the Eigenvalues and Eigenvectors of a Matrix
# a

import numpy as np

# create an array and a vector
A = np.matrix([[4, 0, 1],
              [-1, -6, -2],
               [5, 0, 0]])

v = np.array([1, 2, 3]).reshape(3, 1)

# print the matrix and vector
print('Matrix A:\n', A)
print()
print('Vector v:\n', v)

# Compute the eigenvalues and eigenvectors of the matrix.
evalues, evectors = np.linalg.eig(A)


result = np.dot(A, v)
if np.allclose(result,v * evalues[0]):
  print('\nThe vector is an eigenvector of the matrix')
else:
  print('\nThe vector is not an eigenvector of the matrix')
```

```python
# 2. Principal Component Analysis
# a

import pandas as pd
import matplotlib.pyplot as plt

# read video_game_data.csv into pandas DataFrame
df = pd.read_csv('video_game_data.csv')

# create a scatterplot of user_score and critic_score
plt.scatter(df['user_score'], df['critic_score'])

# add labels and title
plt.xlabel('User Score')
plt.ylabel('Critic Score')
plt.title('User Score vs. Critic Score')

# show the plot
plt.show()
```

```python
#b

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# read video_game_data.csv into pandas DataFrame
df = pd.read_csv('video_game_data.csv')

# seperate user_score and critic_score and reshape to 2D array
score = df[['user_score', 'critic_score']]


# standarize the data
scaler = StandardScaler().fit_transform(score)

# apply PCA
pca = PCA(n_components=2)
xpca = pca.fit_transform(scaler)

# create a new DataFrame with the principal components
xpca_df = pd.DataFrame(xpca, columns=['PC1', 'PC2'])

# add the original data to the new DataFrame
xpca_df['user_score'] = score['user_score']
xpca_df['critic_score'] = score['critic_score']

# plot the new DataFrame
xpca_df.plot.scatter(x='PC1', y='PC2', c='user_score', cmap='coolwarm')

# show the plot
plt.show()

```
